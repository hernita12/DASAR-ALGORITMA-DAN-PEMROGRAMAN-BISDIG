```python
print("=== DATA AKADEMIK MAHASISWA ===")
def identifikasi_data_akademik(data_akademik, kriteria, nilai):
    hasil_identifikasi = []
    for record in data_akademik:
        if kriteria in record and str(record[kriteria]).lower() == nilai.lower():
            hasil_identifikasi.append(record)
    return hasil_identifikasi

# Data Sistem Informasi Akademik (List of Dictionaries)
data_mahasiswa = [
    {"id": "H123", "nama": "Hernita Nur Azizah Lestari", "jurusan": "Bisnis Digital", "mata_kuliah": ["English for Business"]},
    {"id": "H124", "nama": "Kento Aldebaran", "jurusan": "Pendidikan Ilmu Hitam", "mata_kuliah": ["Dasar Pengetahuan Kutukan"]},
    {"id": "H125", "nama": "Zayne Siregar", "jurusan": "Kedokteran", "mata_kuliah": ["Ilmu Bedah"]},
    {"id": "H126", "nama": "Sylus Algazali", "jurusan": "Kriminologi", "mata_kuliah": ["Teori Kriminologi Modern Dasar"]},
    {"id": "H127", "nama": "Rafayel Wiranata Kusuma", "jurusan": "Seni Rupa", "mata_kuliah": ["Dasar Seni Rupa 2D dan 3D"]},
]

# Contoh Penggunaan
print("--- Identifikasi Berdasarkan ID Mahasiswa ---")
id_cari = input("Masukkan ID Mahasiswa yang ingin dicari: ")
hasil_id = identifikasi_data_akademik(data_mahasiswa, "id", id_cari)
if hasil_id:
    for data in hasil_id:
        print(data)
else:
    print(f"Tidak ada data mahasiswa dengan ID: {id_cari}")

print("\n--- Identifikasi Berdasarkan Nama Mahasiswa ---")
nama_cari = input("Masukkan Nama Mahasiswa yang ingin dicari: ")
hasil_nama = identifikasi_data_akademik(data_mahasiswa, "nama", nama_cari)
if hasil_nama:
    for data in hasil_nama:
        print(data)
else:
    print(f"Tidak ada data mahasiswa dengan nama: {nama_cari}")

print("\n--- Identifikasi Berdasarkan Mata Kuliah (Pencarian Parsial) ---")
mk_cari = input("Masukkan Nama Mata Kuliah yang ingin dicari (sebagian atau seluruhnya): ")
hasil_mk = [record for record in data_mahasiswa if any(mk_cari.lower() in mk.lower() for mk in record.get("mata_kuliah", []))]
if hasil_mk:
    for data in hasil_mk:
        print(data)
else:
    print(f"Tidak ada data mahasiswa yang mengambil mata kuliah: {mk_cari}")
```

```python
print("=== APLIKASI E-COMMERCE ===")
produk = {
    1: {"nama": "jeruk", "harga": 17000},
    2: {"nama": "mangga", "harga": 16000},
    3: {"nama": "semangka", "harga": 12000},
    4: {"nama": "alpukat", "harga": 35000},
    5: {"nama": "durian", "harga": 70000},
    6: {"nama": "anggur", "harga": 20000},
    7: {"nama": "stroberi", "harga": 10000},
    8: {"nama": "nanas", "harga": 12000},
    9: {"nama": "apel", "harga": 34000},
    10: {"nama": "jambu", "harga": 17000}
}

keranjang = []

print("=== Selamat datang di Toko Online Chabot ===\n")

while True:
    # Tampilkan daftar produk
    print("Daftar Produk:")
    for kode, item in produk.items():
        print(f"{kode}. {item['nama']} - Rp{item['harga']}")

    try:
        pilihan = int(input("\nMasukkan kode produk yang ingin dibeli: "))
        if pilihan not in produk:
            print("Kode tidak ditemukan!")
            continue
    except ValueError:
        print("Masukkan angka yang valid!")
        continue

    try:
        jumlah = int(input(f"Berapa banyak {produk[pilihan]['nama']} yang ingin dibeli? "))
        if jumlah <= 0:
            print("Jumlah harus lebih dari 0!")
            continue
    except ValueError:
        print("Masukkan angka yang valid!")
        continue

    item = produk[pilihan]
    keranjang.append({
        "nama": item["nama"],
        "harga": item["harga"],
        "jumlah": jumlah,
        "total": item["harga"] * jumlah
    })

    # Tanya lanjut
    lanjut = input("\nMau belanja lagi? (y/n): ")
    if lanjut.lower() != "y":
        break

print("\n=== Struk Belanja ===")
total_bayar = 0
for i, item in enumerate(keranjang, 1):
    print(f"{i}. {item['nama']} x{item['jumlah']} = Rp{item['total']}")
    total_bayar += item['total']

print (f"\nTotal Bayar: Rp{total_bayar}")
print ("Terima kasih telah berbelanja!")
```

```python
print("=== SISTEM RESERVASI HOTEL ===")
class Hotel:
    def __init__(self):
        self.kamar = {
            101: {'tipe': 'Single', 'harga': 500000, 'tersedia': True},
            102: {'tipe': 'Double', 'harga': 750000, 'tersedia': True},
            103: {'tipe': 'Suite', 'harga': 1500000, 'tersedia': True},
        }
        self.reservasi = {}

    def tampilkan_kamar(self):
        print("\nDaftar Kamar:")
        for nomor, info in self.kamar.items():
            status = "Tersedia" if info['tersedia'] else "Tidak Tersedia"
            print(f"Kamar {nomor}: Tipe: {info['tipe']}, Harga: {info['harga']} per malam, Status: {status}")

    def reservasi_kamar(self, nomor_kamar, nama_pelanggan, jumlah_malam):
        if nomor_kamar in self.kamar and self.kamar[nomor_kamar]['tersedia']:
            harga_per_malam = self.kamar[nomor_kamar]['harga']
            total_biaya = harga_per_malam * jumlah_malam
            self.kamar[nomor_kamar]['tersedia'] = False
            self.reservasi[nama_pelanggan] = {
                'nomor_kamar': nomor_kamar,
                'jumlah_malam': jumlah_malam,
                'total_biaya': total_biaya
            }
            print(f"\nReservasi berhasil!")
            print(f"Kamar {nomor_kamar} ({self.kamar[nomor_kamar]['tipe']}) dipesan atas nama {nama_pelanggan} untuk {jumlah_malam} malam.")
            print(f"Total biaya: {total_biaya}")
        else:
            print(f"Kamar {nomor_kamar} tidak tersedia.")

    def batal_reservasi(self, nama_pelanggan):
        if nama_pelanggan in self.reservasi:
            nomor_kamar = self.reservasi[nama_pelanggan]['nomor_kamar']
            self.kamar[nomor_kamar]['tersedia'] = True
            del self.reservasi[nama_pelanggan]
            print(f"Reservasi atas nama {nama_pelanggan} berhasil dibatalkan.")
        else:
            print(f"Tidak ada reservasi atas nama {nama_pelanggan}.")

def main():
    hotel = Hotel()
    while True:
        print("\nMenu:")
        print("1. Tampilkan Kamar")
        print("2. Reservasi Kamar")
        print("3. Batalkan Reservasi")
        print("4. Keluar")
        pilihan = input("Pilih menu (1-4): ")

        if pilihan == '1':
            hotel.tampilkan_kamar()
        elif pilihan == '2':
            nomor_kamar = int(input("Masukkan nomor kamar yang ingin dipesan: "))
            nama_pelanggan = input("Masukkan nama pelanggan: ")
            jumlah_malam = int(input("Masukkan jumlah malam: "))
            hotel.reservasi_kamar(nomor_kamar, nama_pelanggan, jumlah_malam)
        elif pilihan == '3':
            nama_pelanggan = input("Masukkan nama pelanggan untuk membatalkan reservasi: ")
            hotel.batal_reservasi(nama_pelanggan)
        elif pilihan == '4':
            print("Terima kasih! Sampai jumpa.")
            break
        else:
            print("Pilihan tidak valid. Silakan coba lagi.")

if __name__ == "__main__":
    main()
```
